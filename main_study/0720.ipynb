# 프로그래머스 모의고사

```python
def solution(answers):
    supo1 = 0
    supo2 = 0
    supo3 = 0
    supoza1 = [1,2,3,4,5]
    supoza2 = [2,1,2,3,2,4,2,5]
    supoza3 = [3,3,1,1,2,2,4,4,5,5]

    re = 0
    for i in range(len(answers)):

        if i % 5 == 0:
            if answers[i] == supoza1[0]:
                re += 1

        if i % 5 == 1:
            if answers[i] == supoza1[1]:
                re += 1

        if i % 5 == 2:
            if answers[i] == supoza1[2]:
                re += 1

        if i % 5 == 3:
            if answers[i] == supoza1[3]:
                re += 1

        if i % 5 == 4:
            if answers[i] == supoza1[4]:
                re += 1
    supo1 += (re)


    re = 0
    for i in range(len(answers)):

        if i % 8 == 0:
            if answers[i] == supoza2[0]:
                re += 1

        if i % 8 == 1:
            if answers[i] == supoza2[1]:
                re += 1

        if i % 8 == 2:
            if answers[i] == supoza2[2]:
                re += 1

        if i % 8 == 3:
            if answers[i] == supoza2[3]:
                re += 1

        if i % 8 == 4:
            if answers[i] == supoza2[4]:
                re += 1

        if i % 8 == 5:
            if answers[i] == supoza2[5]:
                re += 1

        if i % 8 == 6:
            if answers[i] == supoza2[6]:
                re += 1

        if i % 8 == 7:
            if answers[i] == supoza2[7]:
                re += 1
    supo2 += re

    re = 0
    for i in range(len(answers)):

        if i % 10 == 0:
            if answers[i] == supoza3[0]:
                re += 1

        if i % 10 == 1:
            if answers[i] == supoza3[1]:
                re += 1

        if i % 10 == 2:
            if answers[i] == supoza3[2]:
                re += 1

        if i % 10 == 3:
            if answers[i] == supoza3[3]:
                re += 1

        if i % 10 == 4:
            if answers[i] == supoza3[4]:
                re += 1

        if i % 10 == 5:
            if answers[i] == supoza3[5]:
                re += 1

        if i % 10 == 6:
            if answers[i] == supoza3[6]:
                re += 1

        if i % 10 == 7:
            if answers[i] == supoza3[7]:
                re += 1

        if i % 10 == 8:
            if answers[i] == supoza3[8]:
                re += 1

        if i % 10 == 9:
            if answers[i] == supoza3[9]:
                re += 1
    supo3 += re     
    supo = [supo1, supo2, supo3]
    answer = []
    
    for i,v in enumerate(supo):
        if v == max(supo):
            answer.append(i+1)
        else:
            answer.append(supo.index(max(supo))+1)
    
    answer = list(set(answer))
    
    return answer
```

# 프로그래머스 키패드 누르기

```python
def solution(numbers, hand):
    answer = ''
    Lthumblist= ['*']
    Rthumblist= ['#']


    a = [1,4,7]
    b = [2,5,8,0]
    c = [3,6,9]

    for i in range(len(numbers)):

        if numbers[i] in a:
            answer += "L"
            Lthumblist.append(numbers[i])

        if numbers[i] in c:
            answer += "R"
            Rthumblist.append(numbers[i])

        if numbers[i] in b:
            if numbers[i] == 2:
                v = {2:0,1:1,3:1,5:1,4:2,6:2,8:2,7:3,9:3,0:3,'*':4,'#':4}
                Lv = v[Lthumblist[-1]]
                Rv = v[Rthumblist[-1]]

                if Lv > Rv:
                    answer += "R"
                    Rthumblist.append(numbers[i])
                if Rv > Lv:
                    answer += "L"
                    Lthumblist.append(numbers[i])

                if Rv == Lv:
                    if hand == "left":
                        answer += "L"
                        Lthumblist.append(numbers[i])
                    if hand == "right":
                        answer += "R"
                        Rthumblist.append(numbers[i])


            if numbers[i] == 5:
                v = {5:0,2:1,4:1,6:1,8:1,1:2,3:2,7:2,9:2,0:2,'*':3,'#':3}
                Lv = v[Lthumblist[-1]]
                Rv = v[Rthumblist[-1]]

                if Lv > Rv:
                    answer += "R"
                    Rthumblist.append(numbers[i])
                if Rv > Lv:
                    answer += "L"
                    Lthumblist.append(numbers[i])            

                if Rv == Lv:
                    if hand == "left":
                        answer += "L"
                        Lthumblist.append(numbers[i])
                    if hand == "right":
                        answer += "R"
                        Rthumblist.append(numbers[i])

            if numbers[i] == 8:
                v = {8:0,7:1,5:1,0:1,9:1,2:2,4:2,6:2,1:3,3:3,'*':2,'#':2}
                Lv = v[Lthumblist[-1]]
                Rv = v[Rthumblist[-1]]

                if Lv > Rv:
                    answer += "R"
                    Rthumblist.append(numbers[i])
                if Rv > Lv:
                    answer += "L"
                    Lthumblist.append(numbers[i])  

                if Rv == Lv:
                    if hand == "left":
                        answer += "L"
                        Lthumblist.append(numbers[i])
                    if hand == "right":
                        answer += "R"
                        Rthumblist.append(numbers[i])


            if numbers[i] == 0:
                v = {0:0,8:1,7:2,9:2,5:2,2:3,4:3,6:3,1:4,3:4,'*':1,'#':1}
                Lv = v[Lthumblist[-1]]
                Rv = v[Rthumblist[-1]]

                if Lv > Rv:
                    answer += "R"
                    Rthumblist.append(numbers[i])
                if Rv > Lv:
                    answer += "L"
                    Lthumblist.append(numbers[i])  

                if Rv == Lv:
                    if hand == "left":
                        answer += "L"
                        Lthumblist.append(numbers[i])
                    if hand == "right":
                        answer += "R"
                        Rthumblist.append(numbers[i])

    return answer
```
