# 프로그래머스 [1차] 비밀지도

```python
# 실패한 풀이
def solution(n, arr1, arr2):
    answer = []
    c = ''

    for i in arr1:
        arr1_ = bin(i)
        arr1__ = arr1_[2:]

    for j in arr2:
        arr2_ = bin(j)
        arr2__ = arr2_[2:]

    for k in range(n):
        if arr1__[k] == arr2__[k]:
            c += "#"
        else:
            c += " " 
    print(c)
```

```python
# 정답
def solution(n, arr1, arr2):
    answer = []
    for i in range(0, n):
        answer.append(bin(arr1[i]|arr2[i]).replace('0b', '').zfill(n).replace('1','#').replace('0',' '))
    return answer
```

```python
n = 5
arr1 = [9, 20, 28, 18, 11]
arr2 = [30, 1, 21, 17, 28]
solution(n, arr1, arr2)
```
