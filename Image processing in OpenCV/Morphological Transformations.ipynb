## 1. Erosion

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('j.png',0)
kernel = np.ones((5,5), np.uint8)
erosion = cv.erode(img,kernel,iterations=1)
plt.subplot(121), plt.imshow(img)
plt.subplot(122), plt.imshow(erosion)
```

## 2. Dilation

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('j.png',0)
kernel = np.ones((5,5), np.uint8)
dilation = cv.dilate(img,kernel,iterations=1)
plt.subplot(121), plt.imshow(img)
plt.subplot(122), plt.imshow(dilation)
```

## 3. Opening

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

imgGray = cv.imread('rgbnoise.jpg',0)
kernel = np.ones((5,5), np.uint8)
opening = cv.morphologyEx(imgGray,cv.MORPH_OPEN, kernel)
plt.subplot(121), plt.imshow(imgGray)
plt.subplot(122), plt.imshow(opening)
```
